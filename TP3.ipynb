```python
import numpy as np
import matplotlib.pyplot as plt

# Constante gravitationnelle
g = 9.81

def traj(x, v0, alpha):
    """Calcule la valeur y(x) de la trajectoire."""
    alpha_rad = np.radians(alpha)
    y = -0.5 * g * x**2 / (v0**2 * np.cos(alpha_rad)**2) + x * np.tan(alpha_rad)
    return y

# Paramètres
v0 = 50
xmax = 2500
angles = [20, 45, 60]

# Générer les valeurs de x
x_values = np.linspace(0, xmax, 500)

# Tracer les trajectoires pour les trois angles
plt.figure(figsize=(10, 6))

for alpha in angles:
    y_values = traj(x_values, v0, alpha)
    plt.plot(x_values, y_values, label=f'α = {alpha}°')

# Ajouter des étiquettes et un titre
plt.xlabel('Distance horizontale (x)')
plt.ylabel('Hauteur (y)')
plt.title('Trajectoires pour différentes valeurs de l\'angle de tir')
plt.legend()
plt.grid(True)

# Afficher le graphique
plt.show()
```
